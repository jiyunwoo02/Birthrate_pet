```python
year = ['2017','2018','2019','2020','2021']
male = [12.0,11.6,10.8,9.6,8.6]
female = [11.8,11.5,10.6,9.4,8.5]
plt.rc('font',family = 'Malgun Gothic')
plt.title('혼인율')
plt.plot(year,male,label = '남자', color = 'royalblue')
plt.plot(year,female,label = '여자', color = 'hotpink')
plt.legend()
plt.show()
```
