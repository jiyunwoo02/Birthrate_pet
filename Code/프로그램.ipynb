```python
import tkinter
from tkinter import messagebox
import random

#메인 화면 셋팅
win=tkinter.Tk()
win.title("반려동물 데이터 모아모아")
win.geometry("400x400")
win.resizable(False, False)
        
    
#체크리스트
def new_check():
    check=tkinter.Tk()
    check.title("입양 전 체크리스트")
    check.geometry("400x210")
    check_btn1=Checkbutton(check,text="반려동물 충분히 시간을 할애할 수 있다",pady=5)
    check_btn2=Checkbutton(check,text="반려동물을 입양이 가능한 주거환경에 산다",pady=5)
    check_btn3=Checkbutton(check,text="양육 비용을 감당할 수 있다",pady=5)
    check_btn4=Checkbutton(check,text="끝까지 책임지고 보살펴줄 수 있다",pady=5)
    check_btn5=Checkbutton(check,text="알레르기 테스트를 했다",pady=5)
    check_btn1.pack()
    check_btn2.pack()
    check_btn3.pack()
    check_btn4.pack()
    check_btn5.pack()
    btn70 = Button(check,text="확인",command=click_btn70,width=10)
    btn70.pack()
    
#입양 준비완료
def click_btn70():
        messagebox.showinfo("입양 준비","입양 준비 완료!")   
        
#입양경로 데이터    
def click_btn10():
    way=tkinter.Tk()
    way.title("입양 경로")
    way.geometry("400x400")
  
    
#입양비용 데이터    
def click_btn11():
    cost=tkinter.Tk()
    cost.title("입양 비용")
    cost.geometry("400x400")
    


#입양경로 및 비용    
def new_adopt():
    adopt=tkinter.Tk()
    adopt.title("입양 경로 및 비용")
    adopt.geometry("200x200")
    btn10 = Button(adopt,text="입양 경로", command=click_btn10, width=20)
    btn10.place(x=30, y=45)
    btn11 = Button(adopt,text="입양 비용",command=click_btn11,width=20)
    btn11.place(x=30, y=105)
    
        
        
#이름 추천
def click_btn3():
    random_names=random.randint(0,10)
    newname=names[random_names]
    messagebox.showinfo("이름 추천",newname)

def btncmd():
    messagebox.showinfo("반려동물 등록","반려동물이 정상적으로 등록되었습니다!")
    
def btncmd2():
    messagebox.showinfo("이유","알려주셔서 감사합니다!")    
        
#반려동물 등록    
def new_register():
    register=tkinter.Tk()
    register.title("반려동물 등록")
    register.geometry("400x350")
    txt1 = Text(register, width=30, height=5)
    txt1.pack()
    txt1.insert(END, "종류를 입력해주세요: ")
    txt2 = Text(register, width=30, height=5)
    txt2.pack()
    txt2.insert(END, "이름을 입력해주세요: ")
        
    btn20 = Button(register, text="내 반려동물 등록하기", command=btncmd)
    btn20.pack()
    label = Label(register, text="")
    label.pack()
    label1 = Label(register, text="어떤 이유로 애완동물을 키우게 되었나요?")
    label1.pack()
    chkbox1 = Checkbutton(register, text="반갑게 대해줘서 외롭지 않음")
    chkbox1.pack()
    chkbox2 = Checkbutton(register, text="여가활동을 같이 할 수 있음")
    chkbox2.pack()
    chkbox3 = Checkbutton(register, text="가족 분위기가 좋아짐 ")
    chkbox3.pack()
    chkbox4 = Checkbutton(register, text="기타 ")
    chkbox4.pack()
  
    btn21 = Button(register, text="설문 응답 제출", command=btncmd2)
    btn21.pack()


#로고
lab_p=Label(win)
img=PhotoImage(file="pet.png",master=win)
lab_p.configure(image=img,height=150)
lab_p.pack()


#메인 화면     
btn1 = Button(win,text="입양 전 체크리스트", command=new_check,width=30)
btn1.place(x=90, y=180)
btn2 = Button(win,text="입양 경로 및 비용",command=new_adopt,width=30)
btn2.place(x=90, y=220)
btn3 = Button(win,text="반려동물 이름 추천",command=click_btn3,width=30)
btn3.place(x=90, y=260)
btn4 = Button(win,text="반려동물 등록",command=new_register,width=30)
btn4.place(x=90, y=300)

#이름 더 써 주세요
names=["토토","콩이","코코","초코","나비","닝닝","해피","보리","까미","비니"]



win.mainloop()
```
