```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
import pandas as pd
df = pd.read_excel('borndeath.xlsx', nrows=2, index_col=0)
df
```

```python
df.index.values
```

```python
df.rename(index={'출생아수(명)':'출생아 수','사망자수(명)':'사망자 수'},inplace=True)
df
```

```python
df=df.T
df
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
plt.plot(df.index,df['출생아 수'])
plt.plot(df.index,df['사망자 수'])
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc
plt.subplots(figsize=(12,5))
plt.plot(df.index,df['출생아 수'],marker='o',label='출생아 수')
plt.plot(df.index,df['사망자 수'],marker='o',label='사망자 수')
plt.legend()
plt.title('출생아 수 & 사망자 수')
    
    
```
