```python
import pandas as pd
df = pd.read_excel('assortedfeed.xlsx', index_col=0)
print(df)
```

```python
import matplotlib.pyplot as plt
plt.bar(df['년도'], df['애완개'], label = '애완개')
plt.bar(df['년도'], df['애완고양이'], bottom = df['애완개'], label = '애완고양이')
plt.title('애완동물 배합사료 생산실적')
plt.legend()
```
